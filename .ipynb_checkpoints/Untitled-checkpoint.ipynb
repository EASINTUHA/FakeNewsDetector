```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[_\W]+', ' ', text)  # Remove non-word characters and underscores
    text = re.sub(r"https?://\S+|www\.\S+", "", text)  # Remove URLs
    text = re.sub(r"<.*?>", "", text)  # Remove HTML tags
    text = re.sub(r"[%s]" % re.escape(string.punctuation), " ", text)  # Remove punctuation
    text = re.sub(r"\n", " ", text)  # Remove newlines
    text = re.sub(r"\w*\d\w*", "", text)  # Remove words containing numbers
    return text
```

```python
df["title"]=df["title"].apply(clean_text)
```

```python
df["text"]=df["text"].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Step 1: Initialize separate vectorizers for 'text' and 'title'
text_vectorizer = TfidfVectorizer(max_features=3000)
title_vectorizer = TfidfVectorizer(max_features=1000)

# Step 2: Fit and transform each column
text_tfidf = text_vectorizer.fit_transform(df['text'])
title_tfidf = title_vectorizer.fit_transform(df['title'])

# Step 3: Convert to DataFrames
text_df = pd.DataFrame(text_tfidf.toarray(), columns=[f"text_{w}" for w in text_vectorizer.get_feature_names_out()])
title_df = pd.DataFrame(title_tfidf.toarray(), columns=[f"title_{w}" for w in title_vectorizer.get_feature_names_out()])
```

```python
# Drop original text and title columns
df_other = df.drop(columns=['text', 'title']).reset_index(drop=True)

# Combine all parts into a final DataFrame
df_final = pd.concat([df_other, text_df, title_df], axis=1)
```

```python
# SMOTE Technique : solve imbalance problem (22)
```

```python
# Get count of each class
count_class_0, count_class_1 = df['label'].value_counts()

# Divide by class
df_class_1 = df[df['label'] == 1]  # Minority class (y=1)
df_class_0 = df[df['label'] == 0]  # Majority class (y=0)

# Sample Majority class (y=0) to have the same number of records as the minority class (y=1)
df_class_0_under = df_class_0.sample(count_class_1, random_state=42)

# Join the dataframes containing y=1 and y=0
df_test_under = pd.concat([df_class_0_under, df_class_1])

# Print value counts for the 'label' column to check balance
print(df_test_under['label'].value_counts())

# Print the number of records in the undersampled dataframe
print("Num records = ", df_test_under.shape[0])

# Assign the undersampled dataframe to df
df = df_test_under
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
categorical_columns = df.select_dtypes(include=['object']).columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)
X_encoded = df_encoded.drop('label', axis=1)
y_encoded = df_encoded['label']

best_accuracy = 0
best_random_state = 0

for random_state in range(1, 101):
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=random_state)

    model = RandomForestClassifier(random_state=random_state)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print(f"Best Random State: {best_random_state}")
print(f"Best Accuracy: {best_accuracy}")
```

```python
def roc_auc_plot(y_true, y_proba, label=' ', l='-', lw=1.0):
    from sklearn.metrics import roc_curve, roc_auc_score
    fpr, tpr, _ = roc_curve(y_true, y_proba[:,1])
    ax.plot(fpr, tpr, linestyle=l, linewidth=lw,
            label="%s (area=%.3f)"%(label,roc_auc_score(y_true, y_proba[:,1])))

f, ax = plt.subplots(figsize=(12,8))

roc_auc_plot(y_test,log_model.predict_proba(X_test),label='Logistic Regression ',l='-')
roc_auc_plot(y_test,rf_model.predict_proba(X_test),label='Random Forest Classifier',l='-')
roc_auc_plot(y_test,dt_model.predict_proba(X_test),label='Decision Tree ',l='-')
roc_auc_plot(y_test,knn_model.predict_proba(X_test),label='K Nearest Neighbour ',l='-')
roc_auc_plot(y_test,mlp_model.predict_proba(X_test),label='Multi-Layer Perceptron',l='-')
roc_auc_plot(y_test,et_model.predict_proba(X_test),label='Extra Tree Classifier ',l='-')
#roc_auc_plot(y_test,sgd_model.predict_proba(X_test),label='Stochastic Gradient Descent',l='-')
roc_auc_plot(y_test,gb_model.predict_proba(X_test),label='Gradient Boosting Classifier',l='-')
roc_auc_plot(y_test,ada_model.predict_proba(X_test),label='Adaptive Boosting Classifier',l='-')
roc_auc_plot(y_test,xgb_model.predict_proba(X_test),label='XGboost',l='-')

ax.plot([0,1], [0,1], color='k', linewidth=0.5, linestyle='--',
        )
ax.legend(loc="lower right")
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_title('Receiver Operator Characteristic curves')
sns.despine()
```
